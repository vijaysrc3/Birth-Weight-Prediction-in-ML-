```python
import numpy as np
from numpy.linalg import inv
from time import perf_counter

def linear_regression_closed_form_training(X_train, y_train):
    cpu_time = 0
    betas = []
    
    starting_time = perf_counter()
    
    betas = np.dot(np.linalg.inv(np.dot(X_train.T, X_train)), np.dot(X_train.T, y_train)) 
    
    ending_time = perf_counter()
    cpu_time = ending_time - starting_time
    
    return betas,cpu_time


def linear_regression_closed_form_predict(betas, X_test):
    y_pred = []
    
    y_pred = np.dot(X_test, betas)
       
    return y_pred


def RMSE(y_test, y_pred):
    rmse = -1   
    sum_val = 0
     
    n = len(y_test)
    for i in range(len(y_test)):
        sum_val  = sum_val + (y_test[i]-y_pred[i])**2
    rmse = np.sqrt(sum_val/n)
       
    return rmse
```

```python
def linear_regression_gd_minibatch_training(X_train,y_train,alpha,nEpoch, nIteration, batch_size):
    import random
    from time import perf_counter
    import numpy as np
    random.seed(554433)

    betas = []
    cpu_time = 0
      
    starting_time = perf_counter()  
    betas = np.zeros(X_train.shape[1])
    
    n = len(X_train)
    
    for i in range(nEpoch):
        ind_val = random.randint(0, len(X_train)-batch_size)
        X_train_mb = X_train[ind_val:ind_val+batch_size]
        y_train_mb = y_train[ind_val:ind_val+batch_size]
        
        for j in range(nIteration):
            y_pred = X_train_mb.dot(betas)
            gradient = X_train_mb.T.dot(y_pred - y_train_mb.to_numpy()) / n 

            betas = betas - alpha * gradient

            cost = cost_function(X_train_mb, y_train_mb, betas)

    ending_time = perf_counter()
    cpu_time = ending_time - starting_time
    
      
    return betas,cpu_time
```
